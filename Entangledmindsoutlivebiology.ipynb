```python

import numpy as np
import matplotlib.pyplot as plt

theta_opt = 1.0297  # 59.01
def vitality_score(p00, p11, neuralink_boost=1.17):  # Musk-aligned upload factor
    return 2*p11 + (1 - p00) + (1 - abs(p00 - p11)) * neuralink_boost

p00, p11 = (1 + np.sin(4*theta_opt))/2, (1 - np.sin(4*theta_opt))/2
score = vitality_score(p00, p11)
classical = 1.255
advantage = ((score - classical) / classical) * 100
print(f"QV for Neuralink Upload: {score:.3f} (+{advantage:.1f}%)")

thetas = np.linspace(0, np.pi, 50)
scores = [vitality_score((1 + np.sin(4*t))/2, (1 - np.sin(4*t))/2) for t in thetas]
plt.plot(thetas, scores, 'g-', label='QV Resilience'); plt.axhline(classical, color='r', ls='--', label='Biology Baseline')
plt.scatter(theta_opt, score, c='m', s=200, marker='*', label='59.01 Unbreaking')
plt.xlabel(' (rad)'); plt.ylabel('Coherence Score'); plt.title('QV: Entangled Minds Outlive Biology'); plt.legend(); plt.savefig('qv_musk_proof.png'); plt.show()
```
